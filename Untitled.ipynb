```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, ax = plt.subplots(figsize=(12,5))
plot_acf(yellowstone['Recreation Visits'], lags=30, ax=ax)
plt.show();
```

```python
yellowstone['rec_diff'] = yellowstone['Recreation Visits'].diff()
```

```python
from statsmodels.tsa.arima_model import ARMA
```

```python
ar1 = ARMA(yellowstone.rec_diff.dropna().values, (1, 0)).fit()
```

```python
ar1.summary()
```

```python
plot_acf(yellowstone.rec_diff.dropna().values, lags=10);
```

```python
plot_pacf(yellowstone.rec_diff.dropna().values, lags=10);
```

```python
prediction = ar1.predict()
```

```python
#full_pred = yellowstone['Recreation Visits'].values[0]+np.cumsum(ar1ma1.fittedvalues)

fig, ax = plt.subplots(figsize=(12,5))
ax.plot(date_ticks[1:], yellowstone.rec_diff.dropna().values, lw=2, color='grey', ls='dashed')
ax.plot(date_ticks[1:], prediction, lw=2, color='darkred')
plt.show();
#fig, ax = plt.subplots(figsize=(12,5))
#ax.plot(date_ticks[1:], rec_diff, lw=2, color='grey', ls='dashed',label='Data')
#ax.plot(date_ticks[1:], predictions, lw=2, color='darkred',label='Fit')
#plt.legend()
#plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(yellowstone.rec_diff.dropna().values, prediction)
```
